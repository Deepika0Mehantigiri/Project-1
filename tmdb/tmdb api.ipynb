```python
# Dependencies
import requests
from config import api_key
import pandas as pd
import numpy as np
import csv
from pprint import pprint

#setting year
year = 2022
# setting page number to 1
numb = 1

# base query
# url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={numb}&primary_release_year={year}&sort_by=revenue.desc'"


url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={numb}&sort_by=popularity.desc&include_adult=false&include_video=false&page=1&year=2022&with_watch_providers=netflix&with_watch_monetization_types=flatrate"

#tmdb only displays 20 results per page, and each page must be a separate request, so we 
# must loop the requests by the number of pages in the result
```

```python
# https://api.themoviedb.org/3/discover/movie?api_key=<<api_key>>&language=de-DE&region=DE&release_date.gte=2016-11-16&release_date.lte=2016-12-02&with_release_type=2|3


# https://api.themoviedb.org/3/discover/movie?api_key=<<api_key>>&sort_by=popularity.desc&include_adult=false&include_video=false&page=1&year=2022&with_watch_providers=netflix&with_watch_monetization_types=flatrate

# requesting first page
tmdb = requests.get(url).json()
tmdb
```

```python
# # normalizing JSON, converting to datafram
tmdb_df1 =pd.json_normalize(tmdb, 'results')
tmdb_df= pd.DataFrame(tmdb_df1)
tmdb_df
```

```python
# determining total number of pages in the result
# totalpages = tmdb['total_pages'] + 1
totalpages = 300

# looping through all pages
for numb in range(2, 300):

    # resetting url to new page #
    url = f"https://api.themoviedb.org/3/discover/movie?api_key={api_key}&page={numb}&primary_release_year={year}&sort_by=revenue.desc'"
    
    # advancing counter
    numb += 1

    # requesting next page
    tmdb_page = requests.get(url).json()

    # normalizing JSON, converting to datafram
    tmdb_df2 =pd.json_normalize(tmdb_page, 'results')
    tmdb_page= pd.DataFrame(tmdb_df2)
    tmdb_df = pd.concat([tmdb_df, tmdb_page], ignore_index=True)
    # tmdb_page_list.concat(tmdb_page)



# exporting dataset
pd.DataFrame(tmdb_df).to_csv(f"tmdb_movie_{year}.csv")
```
