```python
%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#CSV path files 

csv_2022="Sources/tmdb_movie_final_2022_budgets.csv"
csv_2021="Sources/tmdb_movie_final_2021_budgets.csv"
csv_2020="Sources/tmdb_movie_final_2020_budgets.csv"
csv_2019="Sources/tmdb_movie_final_2019_budgets.csv"
csv_2018="Sources/tmdb_movie_final_2018_budgets.csv"
csv_2017="Sources/tmdb_movie_final_2017_budgets.csv"

pre_covid="Sources/tmdb_movie_precovid.csv"
covid="Sources/tmdb_movie_covid.csv"
```

```python
tmdb_list_df=pd.DataFrame()
year2 = 2017
for year2 in range(2020, 2023):
    yearlist_df = pd.read_csv(f"Sources/tmdb_movie_final_{year2}_budgets.csv")
    tmdb_list_df = pd.concat([tmdb_list_df, yearlist_df], ignore_index=True)
    year2 += 1
```

```python
# exporting dataset
pd.DataFrame(tmdb_list_df).to_csv(f"Sources/tmdb_movie_covid.csv", encoding='UTF-8')
```

```python
covid_df=pd.read_csv(covid,encoding='UTF-8' )
pre_covid_df=pd.read_csv(pre_covid,encoding='UTF-8' )
pre_covid_df.columns


covid_clean_df=covid_df[['id','original_title','original_language','Country','year','release_date','Genres','popularity','Budget','Revenue','imdb_id']]
pre_covid_clean_df=pre_covid_df[['id','original_title','original_language','Country','year','release_date','Genres','popularity','Budget','Revenue','imdb_id']]
```

```python
#selecting the 10 most popular movie during the covid 2020-2022
most_popular_covid=covid_clean_df.nlargest(10,"popularity")

#make a data frame for the 10 most popular during covid
most_popular_covid_df=pd.DataFrame(most_popular_covid).drop_duplicates()

#selecting the 10 most popular movie before the covid 2017-2019
most_popular_precovid=pre_covid_clean_df.nlargest(10,"popularity")

#make a data frame for the 10 most popular before  covid
most_popular_precovid_df=pd.DataFrame(most_popular_precovid).drop_duplicates()

#selecting the 10 Least popular movie during the covid 2020-2022
least_popular_covid=covid_clean_df.nsmallest(10,"popularity")

#make a data frame for the 10 Least popular during covid
least_popular_covid_df=pd.DataFrame(least_popular_covid).drop_duplicates()

#selecting the 10 Least popular movie before  the covid 2017-2019
least_popular_precovid=pre_covid_clean_df.nsmallest(10,"popularity")

#make a data frame for the 10 Least popular before covid

least_popular_precovid_df=pd.DataFrame(least_popular_precovid).drop_duplicates()
```
